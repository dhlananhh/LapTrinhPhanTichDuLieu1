#### BIỂU ĐỒ ĐƯỜNG (LINE PLOT)

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.available
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
# line chart (biểu đồ đường)

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.figure(figsize=(15, 5))
plt.plot(x, y, "go")
plt.title("Đồ thị")
plt.xlabel("Trục Ox")
plt.ylabel("Trục Oy")

plt.show()
```

```python
# stacked line chart (biểu đồ đường xếp chồng lên nhau)

import numpy as np
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]

x2 = [1, 2, 3, 4]
y2 = [i**3 for i in x2]

plt.plot(x1, y1, "go", x2, y2, 'r^')
plt.title("2 đồ thị")
plt.xlabel("Trục Ox")
plt.ylabel("Trục Oy")

plt.show()
```

```python
# biểu diễn đồ thị 1 dòng 2 cột

import numpy as np
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]

x2 = [1, 2, 3, 4]
y2 = [i**3 for i in x2]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x1, y1, "go")
plt.title("Cột 1")

plt.subplot(1, 2, 2)
plt.plot(x2, y2, 'r^')
plt.title("Cột 2")

plt.suptitle("Biểu diễn đồ thị 1 dòng 2 cột")
plt.show()
```

```python
# biểu diễn đồ thị 2 dòng 1 cột

import numpy as np
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]

x2 = [1, 2, 3, 4]
y2 = [i**3 for i in x2]

plt.figure(figsize=(10, 10))

plt.subplot(2, 1, 1)
plt.plot(x1, y1, "go")
plt.title("Dòng 1")

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r^')
plt.title("Dòng 2")

plt.suptitle("Biểu diễn đồ thị 2 dòng 1 cột")
plt.show()
```

```python
# biểu diễn đồ thị 2 dòng 2 cột

import numpy as np
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]

x2 = [1, 2, 3, 4]
y2 = [i**2 for i in x2]

x3 = [1, 2, 3, 4]
y3 = [i**3 for i in x3]

x4 = [1, 2, 3, 4]
y4 = [i**4 for i in x4]

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

ax[0,0].plot(x1, y1, 'y*')
ax[0,0].set_title('Yellow')

ax[0,1].plot(x2, y2, "go")
ax[0,1].set_title('Green')

ax[1,0].plot(x3, y3, 'r^')
ax[1,0].set_title('Red')

ax[1,1].plot(x4, y4, 'b+')
ax[1,1].set_title('Blue')

plt.suptitle("Biểu diễn đồ thị 2 dòng 2 cột")
plt.show()
```

### BIỂU ĐỒ TRÒN (PIE PLOT)

```python
import matplotlib.pyplot as plt
import numpy as np

names = ["Đại lý A", "Đại lý B", "Đại lý C", "Đại lý D", "Đại lý E"]
amounts = [20,25,15,10,20]
explode = [0,0.1,0,0,0]

plt.figure(figsize=(10,5))
plt.pie(amounts, explode=explode, labels=names, shadow=True, startangle=45)
plt.axis("equal")
plt.legend(title="Danh sách các đại lý", loc='lower left')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
ax.pie(students, labels=langs, autopct='%1.2f%%')

plt.legend()
plt.show()
```

### BIỂU ĐỒ CỘT (BAR CHART)

```python
# biểu đồ cột nằm ngang

import matplotlib.pyplot as plt
import numpy as np

products = ['Sữa', 'Gạo', 'Đường', 'Nước ngọt', 'Muối']
amounts = [70, 82, 60, 55, 40]
_error = [5, 8, 7, 6, 4]

plt.barh(products, amounts, xerr=_error, color='blue')
plt.title('Thống kê doanh số')
plt.xlabel('Doanh số (triệu đồng)')
plt.ylabel('Danh mục')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

products = ['Sữa', 'Gạo', 'Đường', 'Nước ngọt', 'Muối']
amounts = [70, 82, 60, 55, 40]

plt.bar(products, amounts, color='skyblue')
plt.title('Thống kê doanh số')
plt.xlabel('Doanh số (triệu đồng)')
plt.ylabel('Danh mục')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)
ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color', loc='upper right')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)

plt.show()
```

```python
# Grouped bar chart with labels

import numpy as np
import matplotlib.pyplot as plt

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
fig.tight_layout()

plt.show()
```

### BIỂU ĐỒ DẠNG HỘP (BOX PLOT)

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
plt.figure(figsize=(10,10))

ax.set_xlim(0, 160)
ax.set_ylim(0, 50)
ax.set_xlabel('Ngày công')
ax.set_ylabel('Dự án')
ax.set_yticklabels(['Dự án 1', 'Dự án 2', 'Dự án 3'])
ax.grid(True)
ax.broken_barh([(40,50)], (30,8), facecolors=('tab:orange'))
ax.broken_barh([(10,50), (100,20), (130,10)], (20,8), facecolors=('tab:red'))
ax.broken_barh([(20,10), (50,30), (100,50)], (10,8), facecolors=('tab:blue'))

plt.show()
```

```python
# Broken Barh

"""
tạo và hiển thị một biểu đồ dạng "broken barh" 
(biểu đồ cột ngang với các thanh gạch đứt)
"""


import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.broken_barh([(110, 30), (150, 10)], (10, 9), facecolors='tab:blue')
ax.broken_barh([(10, 50), (100, 20), (130, 10)], (20, 9), facecolors=('tab:orange', 'tab:green', 'tab:red'))

plt.show()
```
