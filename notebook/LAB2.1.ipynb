```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(50)
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)
axs[0, 0].plot(x)
plt.show()
```

```python
# scatter plot

import matplotlib.pyplot as plt
x = [1, 2, 3]
y = [1, 4, 9]
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.scatter(x, y, color='g')
ax.set_xlabel('Trục Ox')
ax.set_ylabel('Trục Oy')
ax.set_title('Đồ thị')
plt.show();
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['Đại lý A', 'Đại lý B', 'Đại lý C', 'Đại lý D', 'Đại lý E'])
y = np.array([20, 40, 15, 20, 10])

myExplode = [0,0.2,0,0,0]
myColors = ['blue', 'orange', 'green', 'red', 'purple']
plt.pie(y, labels=x, explode=myExplode, colors=myColors, shadow=True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['Apple','Banana','Chery','Lemon', 'Guava'])
y = np.array([100,200,500,300,600])

myexplode = [0,0.2,0,0,0]

plt.pie(y, labels=x, explode = myexplode)
plt.show()
```
