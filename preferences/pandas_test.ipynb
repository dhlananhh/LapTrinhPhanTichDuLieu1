- map() function is used to transform data (availble for both Series and DataFrame).
- For example : <br> 
<pre>
    original_series = pd.Series([1,2,3,4])
    mapping_strategy = {1: 'one', 2: 'two', 3: 'three', 4: 'four'}
    new_series = original_series.map(mapping_strategy)
</pre>
- Thus, the values stored in the new_series would be ['one', 'two', 'three', 'four'].

```python
mapping_dict = {'A1':'G1','C':'G3','A':'G1','B':'G2'}
df['NHOMKT'] = df['KT'].map(mapping_dict)
df
```

- map() returns an iterator which is not compatible with a pandas series, so we need to use list() to convert it.
- the arguments passed in the map() function must follow that specified order, so that they can be passed to the extraPlus() function orderly correctly.

```python
def extraPlus(x: str, y: float) -> float:
    if((x == 'G1' or x == 'G2') and (y >= 5)):
        return 1
    else:
        return 0

df['DIEMCONG'] = list(map(extraPlus, df['NHOMKT'], df['TBTOAN']))
df
```

- In the case below, you must use the max function provided by NumPy as the estimator. The reason is a little bit tricky, the max itself is a Python built-in function and in certain cases, it can't be as flexible as the np.max function as NumPy was developed to dedicate for complex analysis calculation (especially n-dimensional arrays).
- If I don't use the "hue = 'KT'", the max function still works well because it can easily calculate the maximum value of each KV value (according to the NGONNGU values). But it starts to show its limit when we add the "hue= 'KT'", which is now more complex as for each value of 'KV', we also have sub-categories within it ('KT' in this case), you can see them as 3 two-dimensional arrays, in such a situation, np.max suits  best as it can calculate the maximum value for EACH row within the array.

```python
sns.barplot(x= 'KV', y= 'NGONNGU', hue= 'KT', data= df, estimator= np.max, errorbar= None)
```

# PIE CHART

- 'autopct' stands for 'atuto percentage', it indicates how the percentage figures in your piechart are displayed ('%f1.1%%', '%2.2f%%', ...).

```python
gb = df.groupby('KT').agg(total=('KT', 'count'))

labels = gb.index 
data = list(gb['total'])
colors = sns.color_palette('pastel')

plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()

gb
```
