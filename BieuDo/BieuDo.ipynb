```python
import matplotlib.pyplot as plt
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0) # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
shadow=True, startangle=90)
ax1.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# pie chart

import matplotlib.pyplot as plt
import numpy as np

names = ['Tiền ăn sáng', 'Tiền đi xe bus', 'Tiền ăn vặt', 'Tiền để dành']
amounts = [50, 10, 20, 20]
explode = [0.1, 0, 0, 0]

plt.pie(amounts, explode=explode, labels=names, shadow=True)
plt.axis("equal")
plt.legend(title="Tiền")
plt.title("Chi tiêu 1 tháng của SuSu")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Dữ liệu cho biểu đồ tròn
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]  # Phần trăm tương ứng cho mỗi phần
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']  # Màu cho từng phần

# Tạo biểu đồ tròn
plt.figure(figsize=(6, 6))  # Kích thước biểu đồ
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
# autopct: Hiển thị phần trăm trên biểu đồ, startangle: Góc bắt đầu của biểu đồ

plt.axis('equal')  # Đảm bảo biểu đồ tròn
plt.title('Biểu đồ tròn ví dụ')  # Tiêu đề của biểu đồ

# Hiển thị biểu đồ
plt.show()
```
