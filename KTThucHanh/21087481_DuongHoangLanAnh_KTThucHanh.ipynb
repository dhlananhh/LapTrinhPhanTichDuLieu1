### 7. Điền giá trị thiếu cho biến định lượng của 1 cột nào đó bằng trung vị, trung bình. Xem lại dữ liệu sau khi thay đổi.

```python
# Tạo mới 1 cột dữ liệu có tên là TONGDIEMTOAN
# Giá trị của cột TONGDIEMTOAN = T1 + T2 + T3 + T4 + T5 + T6
df['TONGDIEMTOAN'] = df['T1'] + df['T2'] + df['T3'] + df['T4'] + df['T5'] + df['T6']
```

```python
# Kiểm tra dữ liệu sau khi tạo cột mới là TONGDIEMTOAN
df.head(5)
```

#### 7.1. Điền giá trị thiếu cho biến định lượng của 1 cột nào đó bằng trung vị.

```python
# Tạo mới 1 cột dữ liệu có tên là TrungViTongDiemToan
df['TrungViTongDiemToan'] = None

# Tính trung vị của cột TONGDIEMTOAN
# Sau đó ta sẽ điền giá trị thiếu ở cột TrungViTongDiemToan bằng giá trị trung vị của cột TONGDIEMTOAN

# Tính trung vị của cột TONGDIEMTOAN
trungViDiemToan = df['TONGDIEMTOAN'].median()

# Điền giá trị thiếu ở cột TrungViTongDiemToan bằng giá trị trung vị của cột TONGDIEMTOAN
df['TrungViTongDiemToan'].fillna(trungViDiemToan, inplace=True)
```

```python
# Hiển thị lại dữ liệu sau khi thay đổi
print("Dữ liệu sau khi điền giá trị thiếu bằng trung vị:")
df
```

#### 7.2. Điền giá trị thiếu cho biến định lượng của 1 cột nào đó bằng trung bình.

```python
# Tạo mới 1 cột dữ liệu có tên là TrungBinhTongDiemToan
df['TrungBinhTongDiemToan'] = None

# Tính trung bình của cột TONGDIEMTOAN
# Sau đó ta sẽ điền giá trị thiếu ở cột TrungBinhTongDiemToan bằng giá trị trung bình của cột TONGDIEMTOAN

# Tính trung bình của cột TONGDIEMTOAN
trungBinhDiemToan = df['TONGDIEMTOAN'].mean()

# Điền giá trị thiếu ở cột TrungBinhTongDiemToan bằng giá trị trung bình của cột TONGDIEMTOAN
df['TrungBinhTongDiemToan'].fillna(trungBinhDiemToan, inplace=True)
```

```python
# Hiển thị lại dữ liệu sau khi thay đổi
print("Dữ liệu sau khi điền giá trị thiếu bằng trung bình:")
df
```

### 2. Lọc dữ liệu theo nhiều điều kiện. Sinh viên tự nghĩ câu hỏi và viết lệnh thực thi.

### 3. Tạo 1 biến mới, thực hiện tính giá trị cho biến này theo công thức nào đó. Yêu cầu viết hàm để tính giá trị cho biến này.

```python
# Định nghĩa một hàm để tính giá trị cho biến mới
def tinhTrungBinhDiemToan (row):
    DTB_Toan = (row['T1'] + row['T2'] + row['T3'] + row['T4'] + row['T5'] + row['T6']) / 6
    return DTB_Toan

# Sử dụng hàm apply để tính giá trị và tạo biến mới
df['TrungBinhDiemToan'] = df.apply(lambda row: tinhTrungBinhDiemToan(row), axis=1)
```

```python
# Hiển thị dữ liệu sau khi tạo biến mới
print("Dữ liệu sau khi tạo biến mới:")
df
```
