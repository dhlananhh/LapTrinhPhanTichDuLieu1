# Trực quan hóa dữ liệu bán hàng

## Bài tập 1: Lấy tổng lợi nhuận của tất cả các tháng và hiển thị bằng biểu đồ đường

Biểu đồ đường được tạo ra cần bao gồm các thuộc tính sau:
* Tên trục X = Số tháng
* Tên trục Y = Tổng lợi nhuận


```python
# Import thư viện
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Đọc dữ liệu từ file csv
df = pd.read_csv('../data/sales_data.csv')
```

```python
# Vẽ biểu đồ đường biểu diễn tổng lợi nhuận của tất cả các tháng
# Cách 1: Sử dụng Seaborn

sns.lineplot(x='month_number', y='total_profit', data=df)
plt.xlabel('Số tháng')
plt.ylabel('Tổng lợi nhuận')
plt.title('Biểu đồ đường biểu diễn tổng lợi nhuận theo tháng')
plt.xticks(df['month_number'])
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

## Bài tập 2: Lấy tổng lợi nhuận của tất cả các tháng và hiển thị biểu đồ đường

Biểu đồ đường được tạo ra cần bao gồm:

* Kiểu đường chấm và Màu đường là đỏ
* Hiển thị chú thích ở vị trí dưới bên phải.
* Tên trục X = Số tháng
* Tên trục Y = Số lượng đơn vị đã bán
* Thêm dấu chấm tròn.
* Màu đánh dấu đường là đỏ
* Độ rộng đường là 3

```python
sns.lineplot(x=df['month_number'], y=df['total_profit'], color='r', marker='o', markerfacecolor='k', linestyle='--', linewidth=3, label='Tổng lợi nhuận của tất cả các tháng')
plt.xlabel('Số tháng')
plt.ylabel('Tổng lợi nhuận')
plt.legend(loc='lower right')
plt.title('Biểu đồ đường biểu diễn tổng lợi nhuận theo tháng')
plt.xticks(df['month_number'])
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

## Bài tập 3: Đọc dữ liệu bán hàng của tất cả sản phẩm và hiển thị bằng biểu đồ đa đường (multiline plot)

Hiển thị số lượng đơn vị được bán mỗi tháng cho từng sản phẩm bằng các đường riêng biệt (tức là, mỗi sản phẩm có một đường riêng trên biểu đồ).

```python
sns.lineplot(x=df['month_number'], y=df['facecream'], label='kem dưỡng da', marker='o', linewidth=3)
sns.lineplot(x=df['month_number'], y=df['facewash'], label='sữa rửa mặt', marker='o', linewidth=3)
sns.lineplot(x=df['month_number'], y=df['toothpaste'], label='kem đánh răng', marker='o', linewidth=3)
sns.lineplot(x=df['month_number'], y=df['bathingsoap'], label='xà phòng', marker='o', linewidth=3)
sns.lineplot(x=df['month_number'], y=df['shampoo'], label='dầu gội đầu', marker='o', linewidth=3)
sns.lineplot(x=df['month_number'], y=df['moisturizer'], label='kem dưỡng ẩm', marker='o', linewidth=3)

plt.xlabel('Số tháng')
plt.ylabel('Số lượng đơn vị được bán mỗi tháng của từng sản phẩm')
plt.legend(loc='upper left')
plt.xticks(df['month_number'])
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Biểu đồ đa đường biểu diễn số lượng đơn vị được bán mỗi tháng của từng sản phẩm')
plt.show()
```

## Bài tập 4: Đọc dữ liệu bán hàng của kem đánh răng hàng tháng và hiển thị bằng biểu đồ phân tán (scatter plot)

Ngoài ra, thêm lưới trên biểu đồ. Kiểu đường lưới là –.

```python
sns.scatterplot(x=df['month_number'], y=df['toothpaste'], label='dữ liệu bán hàng của kem đánh răng hàng tháng')
plt.xlabel('Số tháng')
plt.ylabel('Số lượng sản phẩm đã bán')
plt.legend(loc='upper left')
plt.title('Biểu đồ phân tán biểu diễn dữ liệu bán hàng của kem đánh răng hàng tháng')
plt.xticks(df['month_number'])
plt.grid(True, linewidth=1, linestyle='--')
plt.show()
```

## Bài tập 5: Đọc dữ liệu bán hàng của sản phẩm kem dưỡng da và sữa rửa mặt, và hiển thị bằng biểu đồ cột

Biểu đồ cột sẽ hiển thị số lượng đơn vị được bán mỗi tháng cho từng sản phẩm. Thêm một cột riêng biệt cho mỗi sản phẩm trong cùng một biểu đồ.

```python
plt.bar([a-0.25 for a in df['month_number']], df['facecream'], width=0.25, label = 'số lượng kem dưỡng da đã bán', align='edge')
plt.bar([a+0.25 for a in df['month_number']], df['facewash'], width=-0.25, label = 'số lượng sữa rửa mặt đã bán', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(df['month_number'])
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Dữ liệu bán hàng của sản phẩm kem dưỡng da và sữa rửa mặt')
plt.show()
```

## Bài tập 6: Đọc dữ liệu bán hàng của xà phòng tắm trong suốt các tháng và hiển thị bằng biểu đồ cột. Lưu biểu đồ này vào ổ đĩa cứng.

```python
sns.barplot(x=df['month_number'], y=df['bathingsoap'])
plt.xlabel('Số tháng')
plt.ylabel('Số lượng xà phòng đã bán')
plt.title('Dữ liệu bán hàng của xà phòng trong suốt các tháng')
plt.xticks(df['month_number'])
plt.grid(True, linewidth=1, linestyle='--')
plt.savefig('sales_data_of_bathingsoap.png', dpi=150)
plt.show()
```

## Bài tập 7: Lấy tổng lợi nhuận của từng tháng và hiển thị bằng biểu đồ cột dọc (histogram) để xem các khoảng lợi nhuận phổ biến nhất.

```python
labels = ['low', 'average', 'good', 'best']
profit_range = [150000, 175000, 200000, 225000, 250000, 275000, 300000, 325000, 350000]
plt.hist(df['total_profit'], profit_range, label='Dữ liệu lợi nhuận', color='skyblue')
plt.xlabel('Phạm vi lợi nhuận tính bằng đô la')
plt.ylabel('Lợi nhuận thực tế tính bằng đô la')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Dữ liệu lợi nhuận')
plt.show()
```

## Bài tập 8: Tính toán dữ liệu tổng số bán hàng trong năm trước đó cho mỗi sản phẩm và hiển thị bằng biểu đồ tròn (Pie chart)

> Chú ý: Trong biểu đồ tròn, hiển thị số lượng đơn vị được bán hàng trong mỗi năm cho mỗi sản phẩm dưới dạng phần trăm.

```python
pastel_colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#C2C2F0', '#FF99CC']
labels = ['kem dưỡng da', 'sữa rửa mặt', 'kem đánh răng', 'xà phòng tắm', 'dầu gội đầu', 'kem dưỡng ẩm']
salesData = [
    df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
    df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()
]
plt.axis("equal")
plt.pie(salesData, labels=labels, colors=pastel_colors, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Dữ liệu bán hàng')
plt.show()
```

## Bài tập 9: Lấy dữ liệu bán hàng của xà phòng tắm và sữa rửa mặt trong suốt các tháng và hiển thị bằng Subplot (biểu đồ con).

```python
f, axarr = plt.subplots(2, sharex=True)

axarr[0].plot(df['month_number'], df['bathingsoap'], label = 'Dữ liệu bán hàng của xà phòng tắm', color='k', marker='o', linewidth=3)
axarr[0].set_title('Dữ liệu bán hàng của xà phòng tắm')
axarr[1].plot(df['month_number'], df['facewash'], label = 'Dữ liệu bán hàng của sữa rửa mặt', color='r', marker='o', linewidth=3)
axarr[1].set_title('Dữ liệu bán hàng của sữa rửa mặt')

plt.xticks(df['month_number'])
plt.xlabel('Số tháng')
plt.ylabel('Số lượng sản phẩm đã bán')

plt.show()
```

## Bài tập 10: Đọc dữ liệu bán hàng của tất cả sản phẩm và hiển thị bằng biểu đồ xếp chồng (stack plot).

```python
plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(
    df['month_number'],
    df['facecream'],
    df['facewash'],
    df['toothpaste'],
    df['bathingsoap'],
    df['shampoo'],
    df['moisturizer'],
    colors=['m','c','r','k','g','y']
)

plt.xticks(df['month_number'])
plt.xlabel('Số tháng')
plt.ylabel('Số lượng sản phẩm đã bán')
plt.title('Biểu đồ xếp chồng (stack plot) biểu diễn dữ liệu bán hàng của tất cả các sản phẩm theo từng tháng')
plt.legend(loc='upper left')

plt.show()
```
