Import các thư viện

```python
import pandas as pd
```

Đặt tên các cột dữ liệu cần thiết

```python
column_name = ['Id', 'Name', 'Age', 'Weight', 'm0006', 'm0612', 'm1218', 'f0006', 'f0612', 'f1218']
```

3. Tiến hành tải dữ liệu vào chương trình ứng dụng Python và giải quyết vấn đề “Missing header in the csv file”

Đọc dữ liệu lên DataFrame

```python
df = pd.read_csv('../data/patient_heart_rate.csv', names = column_name)
```

Hiển thị 10 dòng đầu tiên

```python
df.head(10)
```

4. Xử lý vấn đề một cột lưu hỗn hợp nhiều dữ liệu, ở đây là cột “Name” chứa bao gồm
“Firstname” và “Lastname”, giải pháp là ta sẽ tách ra làm 2 cột

```python
df[['FirstName', 'LastName']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis=1)
```

Hiển thị 5 dòng đầu tiên

```python
df.head(5)
```

5. Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu. Ta sẽ
chuyển các đơn vị về thành đơn vị chuẩn “kg”

```python
weight = df['Weight']

for i in range(0, len(weight)):
  x = str(weight[i])
  if 'lbs' in x[-3:]:
    x = x[:-3:]
    float_x = float(x)
    y = int(float_x/2.2)
    y = str(y)
    weight[i] = y
  if 'kgs' in x[:-3:]:
    x = x[:-3:]
```

```python
df[['Weight']]
```

Chuyển kiểu dữ liệu của cột "Weight"

```python
# Xoá các ký tự không phải số và không phải dấu thập phân từ cột 'Weight'
df['Weight'] = df['Weight'].str.replace('[^0-9.]', '', regex=True)

# Chuyển đổi kiểu dữ liệu của cột 'Weight' thành float
df['Weight'] = df['Weight'].astype(float)
```

```python
df[['Weight']]
```

```python
df["Weight"].dtype
```

Đổi tên cột Weight thành Weight_kgs

```python
df.rename(columns={'Weight': 'Weight_kgs'}, inplace=True)
```

Hiển thị 5 dòng đầu tiên

```python
df.head(5)
```

6. Vấn đề về xuất hiện dòng dữ liệu rỗng (không có giá trị: NaN). Giải pháp có thể đưa ra là xóa bỏ

```python
df.dropna(how='all', inplace=True)
```

7. Có nhiều dòng dữ liệu bị trùng lắp thông tin hoàn toàn [fullname, lastname, age, weight_kgs,....], giải pháp đưa ra là chỉ giữ lại một dòng dữ liệu, tuy nhiên giải pháp phải dựa trên nghiệp vụ của tập dữ liệu và quan sát của người xử lý.

```python
df = df.drop_duplicates(subset=['FirstName', 'LastName', 'Age', 'Weight_kgs'])
```

```python
df.head(5)
```

8. Xuất hiện dữ liệu bị ảnh hưởng bởi lỗi non-ASCII, không định dạng ASCII. Giải pháp:
Tùy vào nghiệp vụ ta có thể: xóa dữ liệu tại đó, thay thế bằng dữ liệu khác hoặc thay
bằng việc đánh dấu bằng một kí tự khác (ví dụ: ‘warning’)

```python
# Problem 6:
df.FirstName.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
df.LastName.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
print(df)
```

9. Thay giá trị thiếu của tuổi bằng giá trị yếu vị.

```python
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
```

10. Thay giá trị thiếu của cân nặng bằng giá trị trung vị

```python
df['Weight_kgs'].fillna(df['Weight_kgs'].median(), inplace=True)
```

11. “Một cột chứa quá nhiều thông tin cần được phân rã”, như trong bài toán này ta thấy header “m0006” chứa các nội dung bao gồm: m → male, 0006 ~ 00-06 (lần đo chỉ số huyết áp từ 00h- đến 06h). Còn giá trị thì là kết quả huyết áp.

Chúng ta sẽ tách nội dung của cột này ra làm 3 cột sau: PulseRate : giá trị huyết áp, Sex: giới tính ( m: male, f: female) và time: thời gian (từ giờ-đến giờ) như sau:

Bước 1: Tạo melt dữ liệu để có cột gender_time

```python
df = pd.melt(df, id_vars=['Id', 'Age', 'Weight_kgs', 'FirstName', 'LastName'],
             value_name='PulseRate', var_name='gender_time').sort_values(['Id', 'Age', 'Weight_kgs', 'FirstName', 'LastName'])
```

```python
df.head(5)
```

Bước 2: Tạo data frame tạm là kết quả của việc tách cột gender_time

```python
df_temp = df['gender_time'].str.extract("(\D)(\d+)(\d{2})", expand=True)
```

Bước 3: Đặt tên cột cho data frame tạm

```python
df_temp.columns = ['Gender', 'Lower_hour', 'Upper_hour']
```

Bước 4: Nối data frame tạm vào data frame ban đầu

```python
df = pd.concat([df, df_temp], axis=1)
```

```python
df.head(5)
```

Bước 5: Bỏ cột gender_time

```python
df = df.drop(['gender_time'], axis=1)
```

12. Loại bỏ hết các dòng dữ liệu thừa là những dòng có phần PulseRate có dấu -
