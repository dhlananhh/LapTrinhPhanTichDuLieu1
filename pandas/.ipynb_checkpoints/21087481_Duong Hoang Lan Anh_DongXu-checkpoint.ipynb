```python
x = range(4)
y = binom.pmf(x, n, p)
print('x: ', x)
print('y: ', y)
```

```python
# Vẽ scatter
plt.scatter(x, y, color='b')

# Vẽ các đường đứt nét để dễ nhìn
for value in x:
    plt.plot([value, value], [0, y[value]], color='orange', ls='--')

plt.xticks(x)
plt.xlabel('Giá trị')
plt.ylabel('Xác suất')
plt.title('HÀM ĐỘ LỚN PHÂN PHỐI NHỊ THỨC')

plt.show()
```

```python
xs_tich_luy = binom.cdf(x, n, p)
print(xs_tich_luy)
```
