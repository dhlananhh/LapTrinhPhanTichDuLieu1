Import thư viện

```python
import pandas as pd
```

1. Đọc file dữ liệu

```python
thietbi = pd.read_csv('../data/thietbi.csv')
nguoidung_thietbi = pd.read_csv('../data/nguoidung_thietbi.csv')
nguoidung_sudung = pd.read_csv('../data/nguoidung_sudung.csv')
```

2. Vẽ mô hình quan hệ (ERD) từ bảng dữ liệu trên

3. Đổi tên các cột dữ liệu cần thiết

```python
thietbi.rename(columns={'Retail Branding': 'Branding', 'Marketing Name': 'MarketingName'}, inplace=True)
```

```python
thietbi
```

4. Trộn bảng nguoidung_sudung và nguoidung_thietbi thông qua inner join để được bảng kết quả có các cột sau: [outgoing_mins_per_month, outgoing_sms_per_month, monthly_mb, use_id, platform, device]

```python
dfKetQua = pd.merge(nguoidung_sudung, 
                    nguoidung_thietbi[['use_id', 'platform', 'device']],
                    on='use_id')
```

```python
dfKetQua
```

Hoặc dùng left-join tùy theo nhu cầu

```python
# hoặc left-join
dfKetQua = pd.merge(nguoidung_sudung, 
                    nguoidung_thietbi[['use_id', 'platform', 'device']],
                    on='use_id', how='left')
```

```python
dfKetQua
```

5. Tiếp tục inner-join bảng thietbi và dfKetQua để có thêm 2 cột Model và Branding trong bảng kết quả cuối cùng. Lưu ý ở đây khóa liên kết sẽ khác tên

```python
# inner-join
dfKetQua = pd.merge(dfKetQua,
                    thietbi[['Branding', 'Model']],
                    left_on='device',
                    right_on='Model')
```

```python
dfKetQua
```

Hoặc left-join

```python
# left-join
# inner-join
dfKetQua = pd.merge(dfKetQua,
                    thietbi[['Branding', 'Model']],
                    left_on='device',
                    right_on='Model', how='left')
```
