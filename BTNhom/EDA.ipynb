```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#đọc file
df = pd.read_csv('ds_salaries.csv',header=0,delimiter=',',encoding='utf-8')
df.head()
```

```python
#1.Loại bỏ dòng dữ liệu trống
df = df.dropna()
```

```python
#hiển thị số lượng giá trị rỗng
df.isnull().sum()
```

```python
#3.Loại bỏ các dòng dữ liệu trùng nhau
df = df.drop_duplicates()
```

```python
#4.Kiểm tra dữ liệu rỗng
df.isnull().sum()
```

```python
#5.Kiểm tra dữ liệu trùng nhau
df.nunique().to_frame()
```

```python
#6.Đếm số dòng dữ liệu rỗng
df.isnull().sum()
```

```python
#7. Đổi các giá trị trong cột dữ liệu
#7a. Cột Experience Level
#Đổi tên giá trị 'EN', 'MI', 'SE', 'EX' thành 'Entry-level', 'Mid-level', 'Senior-level', 'Expert-level'
df['experience_level'].replace('EN','Entry-level', inplace=True)
df['experience_level'].replace('MI','Mid-level', inplace=True)
df['experience_level'].replace('SE','Senior-level', inplace=True)
df['experience_level'].replace('EX','Expert-level', inplace=True)

df.head()
```

```python
#7b. Cột Employment Type
#Đổi tên giá trị 'PT', 'CT', 'FT', 'FL' thành 'Part-time', 'Contract', 'Full-time', 'Freelance'
df['employment_type'].replace('PT','Part-time', inplace=True)
df['employment_type'].replace('CT','Contract', inplace=True)
df['employment_type'].replace('FT','Full-time', inplace=True)
df['employment_type'].replace('FL','Freelance', inplace=True)

df.head()
```

```python
#7c. Cột Remote Ratio
#Đổi giá trị 0, 50, 100 thành 'Onsite', 'Hybrid', 'Remote'
df['remote_ratio'].replace(0,'Onsite',inplace=True)
df['remote_ratio'].replace(50,'Hybrid',inplace=True)
df['remote_ratio'].replace(100,'Remote',inplace=True)

df.head()
```

```python
#7d. Cột Company Size
#Đổi tên giá trị 'S', 'M', 'L' thành 'Small', 'Medium', 'Large'
df['company_size'].replace('S','Small',inplace=True)
df['company_size'].replace('M','Medium',inplace=True)
df['company_size'].replace('L','Large',inplace=True)

df.head()
```

```python
#9. Hiển thị số lượng giá trị phân biệt của từng cột dữ liệu
df.count()
```

```python
#11. Tách các cột trong bộ dữ liệu ra thành 2 cột: numerical (mang giá trị số) và cột categorical (mang giá trị phân loại)
# Separate columns into numerical and categorical
# Separate columns into numerical and categorical
all_columns = df.columns.to_list()
numerical = ['work_year', 'salary', 'salary_in_usd']
categorical = list(set(all_columns) - set(numerical))

```
