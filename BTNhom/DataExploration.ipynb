Mô tả về dataset

* work_year: Năm trả lương
* experience_level: Mức độ kinh nghiệm trong công việc với các giá trị tương ứng như sau:
    * EN: Entry-level / Junior: Người mới vào nghề
    * MI Mid-level / Intermediate: Người có vài năm kinh nghiệm
    * SE Senior-level/ Expert: Người có kinh nghiệm chuyên môn sâu sắc
    * EX Executive-level / Director: Người có kinh nghiệm lãnh đạo
* employment_type: Loại hình làm việc với các giá trị tương ứng như sau:
    * FT: Full-time: Toàn thời gian
    * PT: Part-time: Bán thời gian
    * CT: Contract: Hợp đồng
    * FL: Freelance: Làm việc tự do
* job_title: Vai trò đã đảm nhiệm trong năm
* salary: Lương thu nhập của nhân viên
* salary_currency: Đơn vị tiền tệ của tiền lương được trả cho nhân viên
* salary_in_usd: Mức lương bằng USD
* employee_residence: Quốc gia cư trú của nhân viên trong năm làm việc
* remote_ratio: Tổng khối lượng công việc được thực hiện từ xa trong năm làm việc, các giá trị tương ứng như sau:
    * 0: Không có công việc từ xa (dưới 20%)
    * 50: Có một phần công việc được làm từ xa
    * 100: Toàn bộ công việc được làm từ xa (hơn 80%)
* company_location: Quốc gia của trụ sở chính hoặc chi nhánh nơi làm việc của nhân viên
* company_size: Số người làm việc cho công ty trong năm, với các giá trị tương ứng như sau:
    * S: ít hơn 50 nhân viên (quy mô nhỏ)
    * M: từ 50 đến 250 nhân viên (quy mô trung bình)
    * L: nhiều hơn 250 nhân viên (quy mô lớn)
